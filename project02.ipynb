📌 코드 설명: 웹 크롤링을 통해 상품 정보 추출

```python
from bs4 import BeautifulSoup
 
headers = {'User-Agent' : 'Mozilla/5.0 (Windows NT 10.0; Win64; x64)AppleWebKit/537.36 (KHTML, like Gecko) Chrome/73.0.3683.86 Safari/537.36'}
data = requests.get('https://minipoe.kr/product/list.html?cate_no=241&',headers=headers)
 
soup = BeautifulSoup(data.text, 'html.parser')

name = soup.select("#anchorBoxId_36539 > div > div.description > p:nth-child(1) > a > span:nth-child(2)")
price = soup.select("#anchorBoxId_36539 > div > div.description > ul > li:nth-child(1) > span:nth-child(2)")
salePrice = soup.select("#anchorBoxId_36539 > div > div.description > ul > li:nth-child(2) > span")
percent = soup.select("#anchorBoxId_36539 > div > div.description > ul > li:nth-child(2) > span > span")

print(name[0].text.strip())
print(price[0].text.strip())
print(salePrice[0].text.split(" ")[0])

if percent:
    print(percent[0].text.strip())  # 할인율만 출력
else:
    print("할인율 없음")
```

👚⭐ 추가 설명
* select(): CSS 선택자 방식으로 HTML 요소를 선택하는 방법
* text.strip(): HTML 태그를 제외한 텍스트만 추출하고, 양쪽 공백을 제거하는 함수
* split(" ")[0]: 텍스트에서 공백을 기준으로 분리하여 첫 번째 항목만 추출 (예: "16,700원"에서 "16,700"만 추출)


📌 코드 설명: 웹 크롤링을 통해 상품 정보 추출 (멀티 페이지 크롤링)

```python
import requests
from bs4 import BeautifulSoup

# 크롤링할 기본 URL (페이지 번호는 `page`로 동적으로 추가됨)
BASE_URL = "https://minipoe.kr/product/list.html?cate_no=241&page="

# HTTP 요청 헤더 (봇 차단 방지)
HEADERS = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/73.0.3683.86 Safari/537.36"
}

# 크롤링한 상품 데이터를 저장할 리스트
data = []

# 페이지 범위 설정 (1~4페이지)
for page in range(1, 5):
    response = requests.get(f"{BASE_URL}{page}", headers=HEADERS)
    
    if response.status_code == 200:
        soup = BeautifulSoup(response.text, 'html.parser')

        # 상품 목록 가져오기 (li 태그 중 id가 'anchorBoxId_'로 시작하는 것 선택)
        items = soup.select("li[id^='anchorBoxId_']")

        for item in items:
            name = item.select("div > div.description > p:nth-child(1) > a > span:nth-child(2)")
            product_name = name[0].text.strip() if name else "상품명 없음"

            price = item.select("div > div.description > ul > li:nth-child(1) > span:nth-child(2)")
            regular_price = price[0].text.strip() if price else "가격 없음"

            salePrice = item.select("div > div.description > ul > li:nth-child(2) > span")
            sale_price = salePrice[0].text.split(" ")[0] if salePrice else "할인 가격 없음"

            percent = item.select("div > div.description > ul > li:nth-child(2) > span > span")
            discount_percent = percent[0].text.strip() if percent else "할인율 없음"


            # 데이터 저장
            data.append({
                "상품명": product_name,
                "정상 가격": regular_price,
                "할인 가격": sale_price,
                "할인율": discount_percent
            })
    
    else:
        print(f"페이지 {page} 요청 실패... (상태 코드: {response.status_code})")

# 크롤링한 데이터 출력
for item in data:
    print(item)

```

4. 크롤링한 데이터 DataFrame으로 출력
- pandas.DataFrame을 활용해 테이블 형태로 데이터 출력

```python
# 📌 데이터프레임으로 변환
df = pd.DataFrame(data, columns=["상품명", "정상 가격", "할인 가격", "할인율"])

# 📌 CSV 파일로 저장
df.to_csv("minipoe_products.csv", index=False, encoding="utf-8-sig")  # 한글 깨짐 방지

print("✅ CSV 파일 저장 완료: minipoe_products.csv")

df
```
