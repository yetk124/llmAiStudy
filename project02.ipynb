```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
from tabulate import tabulate

BASE_URL = "https://minipoe.kr/product/list.html?cate_no=241"

# HTTP 요청 헤더 (봇 차단 방지)
HEADERS = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36"
}

# 크롤링한 상품 데이터를 저장할 리스트
data = []

# 크롤링할 페이지 범위 설정 (1~4페이지)
for page in range(1, 5):
    response = requests.get(f"{BASE_URL}&page={page}", headers=HEADERS)

    # 응답이 정상일 경우만 실행
    if response.status_code == 200:
        soup = BeautifulSoup(response.text, "html.parser")

        items = soup.select("li[id^='anchorBoxId_']")

        for item in items:

            name_spans = item.select("p.name span[style='font-size:13px;color:#000000;']")
            price_spans = item.select("span[style='font-size:12px;color:#000000;']")
            discount_spans = item.select_one("span[style='font-size:12px;color:#a88968;font-style:italic;']")
            
            # 가격을 추출 (판매가, 할인판매가, 할인율 제외)
            itemPrices = [
                price.find(text=True, recursive=False).strip()
                for price in price_spans
                if price.find(text=True, recursive=False) and price.find(text=True, recursive=False).strip() not in ["판매가", "할인판매가"]
            ]
            itemPrice = itemPrices[0] if itemPrices else "가격 없음"
            discount = discount_spans.get_text(strip=True) if discount_spans else "할인 없음"
            
            # "상품명"을 제외한 실제 상품명만 저장
            for span in name_spans:
                text = span.get_text(strip=True)
                if text != "상품명":
                    data.append({"상품명": text, "가격": itemPrice, "할인": discount})

    else:
        print(f"페이지 {page} 요청 실패... (상태 코드: {response.status_code})")

# 데이터프레임 생성 후 출력
df = pd.DataFrame(data)

print(df)
```

```python
from bs4 import BeautifulSoup
 
headers = {'User-Agent' : 'Mozilla/5.0 (Windows NT 10.0; Win64; x64)AppleWebKit/537.36 (KHTML, like Gecko) Chrome/73.0.3683.86 Safari/537.36'}
data = requests.get('https://minipoe.kr/product/list.html?cate_no=241&',headers=headers)
 
soup = BeautifulSoup(data.text, 'html.parser')

name = soup.select("#anchorBoxId_36539 > div > div.description > p:nth-child(1) > a > span:nth-child(2)")
price = soup.select("#anchorBoxId_36539 > div > div.description > ul > li:nth-child(1) > span:nth-child(2)")
salePrice = soup.select("#anchorBoxId_36539 > div > div.description > ul > li:nth-child(2) > span")
percent = soup.select("#anchorBoxId_36539 > div > div.description > ul > li:nth-child(2) > span > span")

print(name[0].text.strip())
print(price[0].text.strip())
print(salePrice[0].text.split(" ")[0])

if percent:
    print(percent[0].text.strip())  # 할인율만 출력
else:
    print("할인율 없음")
```
