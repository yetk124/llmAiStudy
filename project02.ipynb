```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
from tabulate import tabulate

BASE_URL = "https://minipoe.kr/product/list.html?cate_no=241"

# HTTP ìš”ì²­ í—¤ë” (ë´‡ ì°¨ë‹¨ ë°©ì§€)
HEADERS = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36"
}

# í¬ë¡¤ë§í•œ ìƒí’ˆ ë°ì´í„°ë¥¼ ì €ì¥í•  ë¦¬ìŠ¤íŠ¸
data = []

# í¬ë¡¤ë§í•  í˜ì´ì§€ ë²”ìœ„ ì„¤ì • (1~4í˜ì´ì§€)
for page in range(1, 5):
    response = requests.get(f"{BASE_URL}&page={page}", headers=HEADERS)

    # ì‘ë‹µì´ ì •ìƒì¼ ê²½ìš°ë§Œ ì‹¤í–‰
    if response.status_code == 200:
        soup = BeautifulSoup(response.text, "html.parser")

        items = soup.select("li[id^='anchorBoxId_']")

        for item in items:

            name_spans = item.select("p.name span[style='font-size:13px;color:#000000;']")
            price_spans = item.select("span[style='font-size:12px;color:#000000;']")
            discount_spans = item.select_one("span[style='font-size:12px;color:#a88968;font-style:italic;']")
            
            # ê°€ê²©ì„ ì¶”ì¶œ (íŒë§¤ê°€, í• ì¸íŒë§¤ê°€, í• ì¸ìœ¨ ì œì™¸)
            itemPrices = [
                price.find(text=True, recursive=False).strip()
                for price in price_spans
                if price.find(text=True, recursive=False) and price.find(text=True, recursive=False).strip() not in ["íŒë§¤ê°€", "í• ì¸íŒë§¤ê°€"]
            ]
            itemPrice = itemPrices[0] if itemPrices else "ê°€ê²© ì—†ìŒ"
            discount = discount_spans.get_text(strip=True) if discount_spans else "í• ì¸ ì—†ìŒ"
            
            # "ìƒí’ˆëª…"ì„ ì œì™¸í•œ ì‹¤ì œ ìƒí’ˆëª…ë§Œ ì €ì¥
            for span in name_spans:
                text = span.get_text(strip=True)
                if text != "ìƒí’ˆëª…":
                    data.append({"ìƒí’ˆëª…": text, "ê°€ê²©": itemPrice, "í• ì¸": discount})

    else:
        print(f"í˜ì´ì§€ {page} ìš”ì²­ ì‹¤íŒ¨... (ìƒíƒœ ì½”ë“œ: {response.status_code})")

# ë°ì´í„°í”„ë ˆì„ ìƒì„± í›„ ì¶œë ¥
df = pd.DataFrame(data)

print(df)
```
